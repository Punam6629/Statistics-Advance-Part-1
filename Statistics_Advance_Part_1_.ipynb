13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The Central Limit Theorem is one of the most fundamental concepts in statistics. It describes how the distribution of sample means behaves when you repeatedly draw samples from a population.

- Why the CLT is Critical:
- 1 . Foundation for Inferential Statistics:

- CLT allows us to make probabilistic statements about sample means using the normal distribution, even when the population isn't normal.

- 2 . Enables Confidence Intervals & Hypothesis Testing:

- CLT justifies the use of z-tests, t-tests, and confidence intervals, especially when population parameters are unknown.

- 3 . Bridges the Gap Between Data and Theory:

- It explains why sample-based estimates (like averages) are reliable and predictable.

- 4 . Supports Large-Scale Sampling:

- In fields like polling, quality control, and machine learning, the CLT ensures that sample statistics are meaningful reflections of the population.
